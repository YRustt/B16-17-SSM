```python
from numpy import *


def MonteCarlo(f, a, b, n):
    x = random.uniform(a, b, n)
    s = sum(f(x))
    I = (float(b-a)/n)*s
    return I


def f1(x):
    return x * sin(x) * e**x * log(x)

print(MonteCarlo(f1, 1, 3, 1000000))
```

```python
import random as random_number
import matplotlib.pyplot as plt
from numpy import array


def MonteCarloWithError(f, a, b, n, N=100):
    s = 0

    I_values = []
    k_values = []
    for k in range(1, n+1):
        x = random_number.uniform(a, b)
        s += f(x)
        if k % N == 0:
            I = (float(b-a)/k)*s
            I_values.append(I)
            k_values.append(k)
    return k_values, I_values

def f1(x):
    return x * sin(x) * e**x * log(x)

k, I = MonteCarloWithError(f1, 1, 3, 1000000, N=10000)
error = 17.7594 - array(I)

plt.title('Monte Carlo integration')
plt.xlabel('n')
plt.ylabel('error')
plt.plot(k, error)
plt.show()
```

```python
def MonteCarlo2(f, a, b, c, d, n):
    x = random.uniform(a, b, n)
    y = random.uniform(c, d, n)
    s = 0.0
    for i in range(n):
        for j in range(n):
            if (abs(x[i])+abs(y[j]) < 1.0):
                s += f(x[i], y[j])
    #s = sum(f(x, y))
    I = (float(b-a)/n)*(float(d-c)/n)*s
    return I

def f2(x, y):
    return 1.0 * (x**3 + 3*x*y) / e**(-y)

print(MonteCarlo2(f2, -1, 1, -1, 1, 3000))
```
